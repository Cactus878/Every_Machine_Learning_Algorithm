```python
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("doc_text.csv")
data.head(n=10)
```

```python
from category_encoders import OrdinalEncoder
```

```python
encoder = OrdinalEncoder()
encoder.fit(data['label'])
data['label'] = encoder.transform(data['label'])
```

```python
plt.scatter(x=data['left_distance'], y=data['fontsize'], c=data['label'], cmap='viridis')
```

```python
def k_nearest_neighbors(df, unseen_points, k=5):
    results = {"x":[], "y":[], "group":[]}
    for unseen_point in unseen_points:
        ranked_nearest = []
        for index, data in df.iterrows():
            y = data['fontsize']
            x = data['left_distance']
            group = data['label']
            
            dist = math.sqrt((unseen_point[0]- x)**2 + (unseen_point[1] - y)**2)
            ranked_nearest.append((dist,group))
            
        ranked_nearest.sort()
        top_nearest = ranked_nearest[:k]
        
        counts = {}
        for i in top_nearest:
            counts[i[1]] = counts.get(i[1], 0) + 1
        
        nearest_group = max(counts, key=counts.get)
        
        results['x'].append(unseen_point[0])
        results['y'].append(unseen_point[1])
        results['group'].append(nearest_group)
        
    results = pd.DataFrame(results)
    return results
```

```python
unseen_points = [(300, 10), (450, 15), (80, 13), (20, 50)]
k_nearest_neighbors(data, unseen_points)
```
