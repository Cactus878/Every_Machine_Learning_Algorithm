```python
import pandas as pd
import numpy as np
import math
import seaborn as sns
```

```python
data = pd.read_csv('Country-data.csv')
```

```python
data.tail(n=10)
```

```python
def k_centroid_clustering(x_and_y, centroids):
    clustering_results = []
    for pos in x_and_y:
        min_dist = float("inf")
        index = -1
        pos_x = pos[0]
        pos_y = pos[1]


        for c_index, c_pos in centroids.items():
            try:
                pos_x = float(pos_x)
                pos_y = float(pos_y)
            except ValueError:
                continue

            current_euclidean_distance = math.sqrt((pos_x - c_pos[0])**2 + (pos_y - c_pos[1])**2)
        
            if current_euclidean_distance <= min_dist:
                min_dist = current_euclidean_distance
                index = c_index
            
        #var = "%s\t%s\t%s\t%s" % (index, pos_x, pos_y)
        clustering_results.append((index,
                                   pos_x,
                                   pos_y))
        
    return clustering_results

def get_new_centroids(clustering_results, centroids):
    i = 0
    centroid_count = len(centroids)
    current_centroid = 1
    new_centroid = {}
    find_points_from_cluster = lambda data, cluster_id: [t for t in data if t[0] == cluster_id]
    
    while current_centroid <= centroid_count:
        assigned_results = find_points_from_cluster(clustering_results, current_centroid)
        best_x = None
        best_y = None
        best_total_dist = float("inf")

        for current_point in assigned_results:
            total_dist = 0.0
            for point in assigned_results:
                total_dist += math.sqrt((current_point[1] - point[1])**2 + 
                                        (current_point[2] - point[2])**2)
            average_total_dist = total_dist / len(assigned_results)
            if average_total_dist < best_total_dist:
                best_x = current_point[1]
                best_y = current_point[2]
                best_total_dist = average_total_dist
                
        new_centroid[current_centroid] = (best_x, best_y)
        current_centroid += 1
    return new_centroid

def cluster(data, cluster_centers, max_iter):
    i = 0
    results = None
    while i < max_iter:
        print(cluster_centers)
        clustering_results = k_centroid_clustering(data, cluster_centers)
        new_centroids = get_new_centroids(clustering_results, cluster_centers)
        results = clustering_results
        cluster_centers = new_centroids
        i+=1
    return pd.DataFrame(results)
```

```python
get_x_and_y = lambda x, y: list(zip(x, y))
centroids = {1: (200, 174), 2: (200, 0.0659), 3: (0.109,174.0), 4: (0.109,0.0659)}

results = cluster(get_x_and_y(data['exports'], data['imports']), centroids, max_iter=10)
results.columns = ['cluster', 'exports', 'imports']

sns.scatterplot(data=results, x='exports', y='imports', hue='cluster', palette='tab10')
```
